# Introduction to Python Programming Paradigms

```python
# Python is a programming language made by Guido van Rossum and first released in 1991.
# It is a high-level, interpreted language known for its readability and simplicity. Python supports multiple programming paradigms,
# including procedural and object-oriented, thats why its suggested for beginners.
# It has a large standard library and a vibrant ecosystem of third-party packages, making it suitable for a wide range of applications,
# from web development to data science and artificial intelligence.

# Example of procedural programming in Python
def sum(a, b):
    return a + b 
# In procedural programming, we write functions that operate on data, and the data is typically stored in variables.
# It focuses on writing procedures or routines that operate on the data, and the flow of the program is determined by the sequence of function calls.
# Beginners usually start with procedural programming, as its much more simpler at first. But when we create real-world projects
# and application, OOP is recommended as it reduces lines of code and makes it much more organized.

# Example of object-oriented programming in Python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def start_engine(self):
        print(f"The {self.make} {self.model} engine has started.")

my_car = Car("Toyota", "Camry", 2020)
my_car.start_engine()  # Output: The Toyota Camry engine has started.
# In object-oriented programming, we define classes that represent real-world entities, 
# and we create objects (instances) of those classes to represent specific instances of those entities.
# OOP allows us to organize our code in a way that is more intuitive and easier to maintain,
# especially when dealing with complex systems.
```

# Fundamentals of Object-Oriented Programming (OOP)

```python
# Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs.
# It utilizes several techniques from previously established paradigms, including modularity, polymorphism, and encapsulation.
# OOP allows for the creation of classes, which are blueprints for objects, containing attributes (data) and methods (functions).
# This approach helps in organizing code and making it more reusable and maintainable.

# Creating our first class in Python

class Person: # keyword class followed by the name of the class (Person) and a colon
    name = "John Doe" # attributes with a default value
    age = 30
    job = "Software Engineer"
    def introduce(self): # method that introduces the person
        print(f"Hello, my name is {self.name}. I am {self.age} years old and I work as a {self.job}.")

person1 = Person() # to create an instance of the class, we call the class name followed by parentheses.
person1.name = "Saad Sufyan"
person1.age = 15
person1.job = "Data Analyst"
person1.introduce()
person2 = Person()
person2.introduce()
```

```python
# Right now, if we have to modify an instance attr to our choice, it would take one line per attr. But we can utilize the __init__ dunder function,
# which automatically executes when we create our instance. __init__ is a initializer method that allows us to set theinitial state of an object 
# by defining its attributes 

# Using __init__ method to initialize instance attributes

class Point:
    def __init__(self, x=0, y=0): # __init__ method is a special method that is called when an instance of the class is created. It is used to initialize the attributes of the class.
        self.x = x
        self.y = y

p = Point(3, 4) # these parantheses (after the class name) are used to call the __init__ method automatically and create an instance of the class
print(p.x)  # Output: 3
print(p.y)  # Output: 4
```

```python
# Creating a class with a method to calculate the distance from the origin
class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    def distance_from_origin(self):
        return ((self.x **2) +(self.y **2)) **0.5
    
point = Point(4,3)
print(point.distance_from_origin())  # Output: 5.0

# Note : when calling attributes of an instance, we use the dot notation (instance.attribute) and dont specify parentheses, whereas 
# when calling methods of an instance, we use the dot notation (instance.method()) and specify parentheses to indicate that we are calling a function.
```

```python
# Normally, when we print an instance of a class, it will display something like <__main__.ClassName object at memory_address>,
# returning its memory address, which is not very informative. So, we can define a special method called __str__ in our class
# to provide a more meaningful string representation of the instance when we print it. To do this, we can use __str__
# (dunder function starting with double underscores(__)), which gives python the logic on what would the string representation
# of the instance would look like when we print it. 

# Converting an instance to a string using __str__ method

class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    def __str__(self):
        return (f"Point({self.x}, {self.y})")
point = Point(2, 5)
print(point)  # Now python will call the __str__ method to get the string representation of the object
```

```python
# Instances as return values from methods
class Circle:
    def __init__(self, radius=1):
        self.radius = radius
    def area(self):
        return 3.14 * (self.radius ** 2)
    def circumference(self):
        return 2 * 3.14 * self.radius
print(Circle(5).area())  # Output: 78.5
print(Circle(5).circumference())  # Output: 31.400000000000002
```

```python
# OOP Real World Application: SMS Store
# Create a new class, SMS_store. The class will instantiate SMS_store objects, similar to
# an inbox or outbox on a cellphone:
# my_inbox = SMS_store()
# This store can hold multiple SMS messages (i.e. its internal state will just be a list of
# messages). Each message will be represented as a tuple:
# (has_been_viewed, from_number, time_arrived, text_of_SMS)
# The inbox object should provide these methods:
# my_inbox.add_new_arrival(from_number, time_arrived, text_of_SMS)
# # Makes new SMS tuple, inserts it after other messages
# # in the store. When creating this message, its
# # has_been_viewed status is set False.
# my_inbox.message_count()
# # Returns the number of sms messages in my_inbox
# my_inbox.get_unread_indexes()
# # Returns list of indexes of all not-yet-viewed SMS messages
# my_inbox.get_message(i)
# # Return (from_number, time_arrived, text_of_sms) for message[i]
# # Also change its state to "has been viewed".
# # If there is no message at position i, return None
# my_inbox.delete(i) # Delete the message at index i
# my_inbox.clear() # Delete all messages from inbox
# Write the class, create a message store object, write tests for these methods, and implement
# the methods.
 
```

```python
# Response:
class SMS_store:
    def __init__(self):
        self.inbox = []
    def __str__(self):
        return (f"{self.inbox}")
    def add_new_arrival(self, from_number, time_arrived, text_of_SMS):
        has_been_viewed = False
        message = (
            has_been_viewed,
            from_number,
            time_arrived,
            text_of_SMS
    )
        self.inbox.append(message)
    def message_count(self):
        total_messages = len(self.inbox)
        return total_messages
    def get_unread_indexes(self):
        unread_indexes = []
        for index, item in enumerate(self.inbox):
            if item[0] == False:
                unread_indexes.append(index)
        return unread_indexes
    def get_message(self, i):
        if i < len(self.inbox):
            message = self.inbox[i]
            self.inbox[i] = (True, message[1], message[2], message[3])
            return (message[1], message[2], message[3])
        else:
            return None
    def delete(self, i):
        if i < len(self.inbox):
            del self.inbox[i]
    def clear(self):    
        self.inbox.clear()

my_inbox = SMS_store()
my_inbox.add_new_arrival("0312", "10 PM", "Hello")
my_inbox.add_new_arrival("0322", "12 PM", "We will be coming late")
my_inbox.add_new_arrival("0314", "1 AM", "Open the door")
print(my_inbox)
print("Total Messages in inbox :",my_inbox.message_count())

print(my_inbox.get_message(0))
print(my_inbox)
print("Total Messages in inbox :",my_inbox.message_count())

print("Indexes of unread messages in inbox :",my_inbox.get_unread_indexes())

my_inbox.delete(0)
print(my_inbox)

my_inbox.add_new_arrival("0321", "11 PM", "Where are you?")
print(my_inbox)

my_inbox.clear()
print(my_inbox)
```
