```python
# Creating our first class in Python
class Person:
    name = "John Doe"
    age = 30
    job = "Software Engineer"
    def introduce(self):
        print(f"Hello, my name is {self.name}. I am {self.age} years old and I work as a {self.job}.")

person1 = Person()
person1.name = "Saad Sufyan"
person1.age = 15
person1.job = "Data Analyst"
person1.introduce()
person2 = Person()
person2.introduce()
```

```python
# Using __init__ method to initialize attributes to factory default values
class Point:
    def __init__(self):
        self.x = 0
        self.y = 0

p = Point()
print(p.x)  # Output: 0
print(p.y)  # Output: 0
```

```python
# Using __init__ method to initialize attributes with custom values
class Student:
    def __init__(self, name = "Unknown", age = 0, grade = "N/A", roll_number = "N/A"):
        self.name = name
        self.age = age
        self.grade = grade
        self.roll_number = roll_number

student1 = Student("Alice", 20, "A", "12345")
print(student1.name)  # Output: Alice
print(student1.age)   # Output: 20
print(student1.grade) # Output: A
print(student1.roll_number) # Output: 12345
student2 = Student()
print(student2.name)  # Output: Unknown
print(student2.age)   # Output: 0
print(student2.grade) # Output: N/A
print(student2.roll_number) # Output: N/A
```

```python
# Creating a class with a method to calculate the distance from the origin
class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    def distance_from_origin(self):
        return ((self.x **2) +(self.y **2)) **0.5
    
point = Point(4,3)
print(point.distance_from_origin())  # Output: 5.0
```

```python
# Converting an instance to a string using __str__ method
class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    def __str__(self):
        return (f"Point({self.x}, {self.y})")
point = Point(2, 5)
print(point)  # Now python will call the __str__ method to get the string representation of the object
```

```python
# Instances as return values from methods
class Circle:
    def __init__(self, radius=1):
        self.radius = radius
    def area(self):
        return 3.14 * (self.radius ** 2)
    def circumference(self):
        return 2 * 3.14 * self.radius
print(Circle(5).area())  # Output: 78.5
print(Circle(5).circumference())  # Output: 31.400000000000002
```

```python
# Exercise:
# Create a new class, SMS_store. The class will instantiate SMS_store objects, similar to
# an inbox or outbox on a cellphone:
# my_inbox = SMS_store()
# This store can hold multiple SMS messages (i.e. its internal state will just be a list of
# messages). Each message will be represented as a tuple:
# (has_been_viewed, from_number, time_arrived, text_of_SMS)
# The inbox object should provide these methods:
# my_inbox.add_new_arrival(from_number, time_arrived, text_of_SMS)
# # Makes new SMS tuple, inserts it after other messages
# # in the store. When creating this message, its
# # has_been_viewed status is set False.
# my_inbox.message_count()
# # Returns the number of sms messages in my_inbox
# my_inbox.get_unread_indexes()
# # Returns list of indexes of all not-yet-viewed SMS messages
# my_inbox.get_message(i)
# # Return (from_number, time_arrived, text_of_sms) for message[i]
# # Also change its state to "has been viewed".
# # If there is no message at position i, return None
# my_inbox.delete(i) # Delete the message at index i
# my_inbox.clear() # Delete all messages from inbox
# Write the class, create a message store object, write tests for these methods, and implement
# the methods.
 
```

```python
# Response:
class SMS_store:
    def __init__(self):
        self.inbox = []
    def __str__(self):
        return (f"{self.inbox}")
    def add_new_arrival(self, from_number, time_arrived, text_of_SMS):
        has_been_viewed = False
        message = (
            has_been_viewed,
            from_number,
            time_arrived,
            text_of_SMS
    )
        self.inbox.append(message)
    def message_count(self):
        total_messages = len(self.inbox)
        return total_messages
    def get_unread_indexes(self):
        unread_indexes = []
        for i in self.inbox:
            if i[0] == False:
                unread_indexes.append(i)
        return unread_indexes
    def get_message(self, i):
        if i < len(self.inbox):
            message = self.inbox[i]
            self.inbox[i] = (True, message[1], message[2], message[3])
            return (message[1], message[2], message[3])
        else:
            return None
    def delete(self, i):
        if i < len(self.inbox):
            del self.inbox[i]
    def clear(self):    
        self.inbox.clear()

my_inbox = SMS_store()
my_inbox.add_new_arrival("0312", "10 PM", "Hello")
print(my_inbox)
print("Total Messages in inbox :",my_inbox.message_count())
print(my_inbox.get_message(0))
print(my_inbox)
print("Indexes of unread messages in inbox :",my_inbox.get_unread_indexes())
my_inbox.delete(0)
print(my_inbox)
my_inbox.add_new_arrival("0321", "11 PM", "Where are you?")
print(my_inbox)
my_inbox.clear()
print(my_inbox)
```
