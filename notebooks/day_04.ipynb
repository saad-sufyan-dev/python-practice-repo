# Exception Handling in python

Generally, when we execute a python program, we use a python interpreter (it translates code to binary line by line) as python is an interpreted  
language, unlike C and C++.  

When a python interpreter encounters an error while interpreting (such as converting a string into an integer), it returns to the program and  
checks if user has provided what to do if such error occured (how to encounter ValueError).  
If yes, it handles error accordingly, otherwise, it crashes and terminates flow of program.  

For Example, if we have 10 lines of code and error happens at line 5, then flow of program will terminate at line 5.  
Thats why its imp to learn how to handle exceptions in python.

````python
# Using try and except 
try:
    a = int(input("Enter an integer: "))
    print(f"Multiplication Table of {a} is:")
    for i in range(1, 11):
        print(f"{a} X {i} = {a * i}")
except ValueError:
    print("Input not an integer!")

print("\nSome imp lines of code.")
# If we enter a number, code will execute normally. But if someone enters a string, instead of crashing the program the interpreter would 
# handle that exception (printing that input is invalid) and successfully execute whole program without halting in the middle.
````

````python
# In python we can also raise errors and halt the program execution, like when a situation breaks real-life logic,
# even if it doesnt trigger a standard system crash. 

# Using raise keyword

def withdraw_money(balance, amount):
    if amount <= 0:
        raise ValueError("Withdrawal amount must be greater than zero.")
    
    if amount > balance:
        # Raising an error because you cannot withdraw more than you own
        raise PermissionError("Insufficient funds for this transaction.")
    
    return balance - amount

# Executing the code
try:
    current_balance = 5000
    # Attempting to withdraw more than available balance
    current_balance = withdraw_money(current_balance, 6000)
except PermissionError as error: # --> here, error is an alias of PermissionError
    print(f"Transaction Denied: {error}")
except ValueError as e:
    print(e)

print("\n Some imp lines of code") # --> If we hadnt handled the exception/error using try-except block, the flow had been halted at line of error.
````

```text
[ Code Execution ]
 └── Call: withdraw_money(5000, 6000)
      ├── Condition Check: 6000 > 5000 (True)
      └── raise PermissionError ──> (Execution Halts)
                                           │
[ Exception Handling ]                     ▼
 └── caught by 'except PermissionError as error'
      └── Prints: "Transaction Denied: Insufficient funds for this transaction."

````python
a = input("Enter a number: ")
try: 
    a = int(a)
    b = 1/a
    print(f"1 divided by {a} is {b}")

except ZeroDivisionError: 
    print("Cant divide by 0!")

except ValueError:
    print("Invalid input, not an integer")

except Exception as e: # here storing any other error in an alias 'e' and handling it using except 
    print(f"Some error occured:\n{e}")
````

````python
# the finally clause in try-except block: it always runs whether an error occurs or not
# Common use-case: Closing a file

file = None
try:
    file = open("data.txt")
    data = file.read()
except FileNotFoundError:
    print("File not found")
finally:
    if file != None:
        file.close()

# Usually we use 'with' syntax, which automatically closes file
try:
    with open("data.txt") as f:
        data = f.read()

except FileNotFoundError:
    print("File not Found")
````

# Statements (break, continue and pass) in python

````python
# break statement terminates the loop completely and transfers flow of execution to the next statement outside the loop
# Usually used with for-else statement, where the else block belongs to for loop instead of if statement/
numbers = [1,2,3,4]

for num in numbers:
    if num == 4:
        print("Found")
        break # --> If val is found, 'break' will terminate the loop, and the else block will be skipped

else: # --> If no break statement executes during looping, then else block will execute 
    print("Not Found")
````

````python
# continue statement skips remaining code inside current iteration inside the loop and transfers flow directly to the next iteration
try:
    a = int(input("Enter a number:"))
    for i in range(11):
        if i == 0:
            print("Skip the iteration")
            continue # --> continue will skip rem code under current iteration 
        print(f"{a} X {i} = {a*i}")
except:
    print("Some error occured")
````
