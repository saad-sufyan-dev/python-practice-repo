# Introduction to File I/O in Python

I/O --> Input/Output, reading and writing data to files.

Volatile v/s Non-volatile Storage:
1. RAM --> Volatile(temporary), which is where active runtime data, program execution states and variables of a program are stored,  
    but data is lost when the power is turned off.
2. Hard Disk/SSD --> Non-volatile, which retains data even when the power is turned off.

We generally use files to store data on the hard disk/SSD, which allows us to save data permanently and access it later.

````python
# To open a file using python, we can use the built-in `open()` function.
f = open("demo.txt", "r")  # "r" stands for read mode
content = f.read()  # Read the entire content of the file
print(content)  # Print the content of the file

f.close()  # Close the file after reading
````

````python
# Using readline() to read a file line by line
b = open("demo.txt")
line1 = b.readline()
line2 = b.readline()
print(f"This is line 1 of demo.txt: {line1}") # Space will come after line one because of hidden \n (used for next line)
print(f"This is line 2 of demo.txt: {line2}")
b.close()
````

````python
# Using "w" mode (overwrite) and "a" mode (appending)

f = open("demo.txt","w")
f.write("I will learn file handling in python tomorrow.")
f.close()

b = open("demo.txt","a")
b.write("\nThen I'll upload it to Github.")
b.close()
````

````python
# If u open a file that doesnt exist using w or a mode, python automatically creates it for u
f = open("sample.txt","a")
f.write("This file was created automatically using 'a' mode")
f.close()
````

```text
Meaning of different combinations of modes:
"r" --> read mode (default), stream is positioned at the beginning of the file, throws error if file doesn't exist
"w" --> write mode, stream is positioned at the beginning of the file and truncates the file to zero length, creates file if it doesn't exist
"a" --> append mode, stream is positioned at the end of the file, creates file if it doesn't exist
"r+" --> read and write mode, stream is positioned at the beginning of the file, throws error if file doesn't exist
"w+" --> write and read mode, stream is positioned at the beginning of the file and truncates the file to zero length, creates 
file if it doesn't exist
"a+" --> append and read mode, stream is positioned at the end of the file, creates file if it doesn't exist.


````python

# Using w+ mode
f = open("demo.txt","w+")
print(f.read()) # return nothing as file is truncated
f.write("Python is an amazing language.")
f.close()

# Using a+ mode
f = open("demo.txt","a+")
f.write("\nBecause of its readability.")
print(f.read()) # return nothing as pointer is placed at the end of file
f.close()

# Using r+ mode
f = open("demo.txt", "r+")
f.write("It also has a vast amount of libraries to help u solve problems")
print(f.read()) # return nothing as pointer is placed at the end of file after overwriting, need to use f.seek(0) to move pointer to the beginning.
f.close()
````

````python
# Using `with` statement to open a file, which automatically takes care of closing the file after the block of code is executed.
with open("demo.txt","r") as f:
    content = f.read()
    print(content)
````

````python
# Using os module to delete a file

# A module is like a code library that contains pre-written code to perform specific tasks, and we can import 
# and use these predefined functions in our own code written by others to save time and effort.
# Some modules come preinstalled in python, including os.

import os
os.remove("sample.txt")
````

<h3> File I/O Practice Problem: </h3>

Create a new file “practice.txt” using python. Add the following data in it:

<p style="color:red;">Hi everyone<br>
we are learning File I/O<br>
using Java.<br>
I like programming in Java.</p>

Replace all occurrences of “java” with “python” in above file.

WAF that searches if the word “learning” exists in the file or not.

````python
with open("practice.txt","w+") as f:
    f.write("Hi everyone\nwe are learning File I/O\nusing Java\nI like programming in Java")
    f.seek(0)
    content = f.read()
    new_content = content.replace("Java", "Python") # There is already built in method in str class which replaces values
    f.seek(0)
    f.write(new_content)

def check_for_word(data, word):
    if(word in data):
        print("found")
    else:
        print("not found")

check_for_word(new_content, "learning")
````

# Revisiting Data Types in Python

````python
# We already know how strings, integer, float and boolean data types work in python. 
a = 10 # integer
b = 3.14 # float
c = "Hello" # string
d = True # boolean

# All of them are immutable data types, which means that their values cannot be changed after they are created.
````

````python
# Now we will learn about some other data types in python, which are list, dictionary, set and tuple.

# List is a collection of items which is ordered (indexing) and mutable (changeable), and allows duplicate members. It is defined using square brackets [].
a = ["apple", "banana", "cherry", "apple"]
print(a)
print(a[2])
a[3] = "orange" # changing value at index 3
print(a)

# Tuples are similar to lists, but they are immutable (unchangeable) and defined using parentheses ().
b = ("apple", "banana", "cherry", "apple")
print(b)
print(b[2])
# b[3] = "orange" # this will throw an error as tuples are immutable

print(type(a)) # <class 'list'>
print(type(b)) # <class 'tuple'>
````

````python
# Dictionaries are a collection of key-value pairs which is unordered (no indexing) and mutable (changeable), and does not allow duplicate members. 
# Its defined using curly braces {}.
c = {"name": "Alice", "age": 25, "city": "New York", "is_adult": True, "children": ["Bob", "Charlie"]}
print(c)
print(c["name"])
c["age"] = 26 # changing value of key "age"
print(c)
# In dictionaries, we can have values of any data type, including lists and even other dictionaries. But keys must be of immutable data type,
# which means that we can use strings, integers, floats and tuples as keys, but not lists or dictionaries.
````

````python
# Nested Dictionary: 
# We can also convert our value to a dictionary

info = {
    "name": "Saad Sufyan",
    "subjects": {
        "physics": 89,
        "chem": 95,
        "math": 100
    }
}

print(info["subjects"])
print(info["subjects"]["physics"]) # to access value of key "physics" inside the nested dictionary "subjects"

null_dict = {} # empty dictionary
print(null_dict)
print(type(null_dict))
````

````python
# Sets is the collection of unique items which is unordered and immutable, and does not allow duplicate members. It is defined using curly braces {}.
collection = {"apple", "banana", "cherry", "apple",1,2,3,1,2,34,5}
print(collection) # it will only print unique values, and order is not guaranteed

null_set = set() # empty set
print(null_set)
print(type(null_set))
````

Immutable data types: string, integer, float, boolean, tuple, set --> value can be overwritten but not modified<br>
Mutable data types: list, dictionary
