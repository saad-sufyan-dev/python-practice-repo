# Functions and Recursions

```python
# A functions is a block of code that performs a specific task. It can take inputs, called parameters, and can return an output.
#  Functions are reusable, which means you can call them multiple times in your code without having to rewrite the same code.
def sum(a, b): # func parameters
    return a + b # return statement gives data back to the caller

result = sum(3, 5) # func arguments
print(result) # Output: 8
```

```python
# Function Application: Dollar to PKR converter
def converter(d, rate=280):
    r = d*rate
    print(f"{d} USD = {r} PKR")

converter(100)
```

```python
# Recursive function: A function that calls itself repeatedly in order to solve a problem. 
# It typically has a base case that stops the recursion and a recursive case that breaks the problem into smaller subproblems.
def show(n):
    if (n <=0): # base case
        return
    print(n)
    show(n-1) # recursive case
    print("END") # will be printed after the recursive call is completed and function starts to return back up the call stack

show(5) # Output: 5 4 3 2 1 END END END END END

# Note: Recursion is similiar to loops, but for beginners it can be a bit tricky to understand how it works, so loops are generally
# recommended for beginners to solve problems, and once you are comfortable with loops, you can start exploring recursion as well.
```

```python
# Recursion Application: Factorial of a number
def factorial(n):
    if (n==1 or n==0):
        return 1
    elif (n>1):
        return n * factorial(n-1)
    else:
        return None

a = factorial(5)
print(a)

# This is the mental model that makes recursion click:
# Going down: "I don't know the answer yet, so I'll ask a smaller version of the same problem."
# Coming back up: "Now that the smaller problem gave me an answer, I'll use it."
```

# Advanced Concepts of Object Oriented Programming 

```python
# OOP is based on 4 main principles:
# 1. Encapsulation: The bundling of data(attributes) and methods within a single unit, such as a class, and controlling access to that data.
# It restricts direct access to some of the object's components (such as private attributes), which can prevent the accidental modification of data.

# 2. Abstraction: The process of hiding the complex implementation details of a system and exposing only the necessary features to the user.
# 3. Inheritance: The mechanism by which one class can inherit properties and behaviors (attributes and methods) from another class.
# This promotes code reusability and establishes a natural hierarchical relationship between classes.

# 4. Polymorphism: the same method works differently depending on the object that uses it
```

```python
# To understand these principles, let's take an example of a simple class called "Car".

# 1. Encapsulation: The Car class encapsulates the attributes (model, color, engine_status) and methods (start_engine) that
# operate on those attributes within a single unit (the Car class). The engine_status attribute is private, which means it can only be accessed
# and modified within the Car class, and cannot be accessed directly from outside the class.

# 2. Abstraction: The Car class abstracts away the complex details of how the engine works and provides a simple interface 
# (the start_engine method) for the user to interact with the car. The user doesn't need to know how the engine works, they just need to know how to start it

# 3. Inheritance: We can create a new class called "SportsCar" that inherits from the Car class (base/parent class). The SportsCar class can reuse
# the attributes and methods of the Car class, and can also add its own unique attributes and methods.

# 4. Polymorphism: The start_engine method can be overridden in the SportsCar class to provide a different implementation that is specific to
# sports cars. This allows us to use the same method name (start_engine) for different types of cars, and the correct implementation will 
# be called based on the type of car (or object) we are working with.

class Car:
    def __init__(self, model, color):
        self.model = model # public attribute
        self.__engine_status = "off" # private attribute --> can only be used within the class, cannot be accessed directly from outside the class
        self.color = color # public attribute
    def start_engine(self):
        self.__engine_status = "on"
        print(f"{self.model}'s engine started.")

class SportsCar(Car):
    def __init__(self, model, color):
        super().__init__(model, color) # super() is used to call the __init__ or other methods of the parent class (Car).
    def start_engine(self):
        self.__engine_status = "on"
        print(f"Your {self.color} {self.model} roars to life")

class ElectricCar(Car):
    def __init__(self, model, color):
        super().__init__(model, color)
    def start_engine(self):
        self.__engine_status = "on"
        print(f"Your {self.color} {self.model} starts silently")

Car1 = Car("Toyota","Blue")
Car1.start_engine()
Car2 = ElectricCar("Tesla","Green")
Car2.start_engine()
Car3 = SportsCar("Ferarri","Red")
Car3.start_engine()
# print(Car1.__engine_status) # This will raise an error because __engine_status is a private attribute and cannot be accessed directly from outside the class.
```

```python
# Now we will create a simple class called "Account" and implement all 4 prionciples of OOP in it. 

# There are 2 types of attr:
# Class attr --> when an attr is supposed to be same in every object within a class, such as schoolname in Student class, we
# store it in a class attr, instead of storing it as an instance attr as it will consume more memory.
# Instance attr --> when an attr, such as studentname or marks, are not the same for every obj. We store these attr in obj (memory).

class Account:
    def __init__(self, username, password, inidep=0):
        self.username = username
        self.inidep = inidep
        self.__acc_pass = password # Private attr
        self.balance = inidep
    
    def credit(self, val):
        self.balance += val
        print(f"Your Account Balance is : {self.balance}")

    def debit(self, val):
        self.balance -= val
        print(f"Your Account Balance is : {self.balance}")

    def check_balance(self):
        print(f"Your Account Balance is : {self.balance}")

    def reset_password(self, new_password):
        self.__acc_pass = new_password

class SavingsAccount(Account):
    def __init__(self, username, password, inidep=0, interest_rate=5):
        self.interest_rate = interest_rate
        super().__init__(username, password, inidep)

    def add_interest(self):
        interest = self.balance*(self.interest_rate/100)
        self.balance += interest
        print(f"Your Account Balance is : {self.balance}")


       


a1 = Account("Ali_Asghar","abc123",32000)
a2 = SavingsAccount("Umer_Affan","hfs@0987",100000)

a1.credit(50000) 
a1.debit(25000)

a2.add_interest()
```

```python
# A decorator is a function that takes another function as an argument and extends its behavior without explicitly modifying it.

# We can use the @property decorator to covnert a method into a property, which allows us to access it like an attribute, without needing to call it as a method.
# It also calculates this attr val evry time it is accessed, which means it will always return the most up-to-date value.

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    @property
    def grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 80:
            return "B"
        elif self.marks >= 70:
            return "C"
        elif self.marks >= 60:
            return "D"
        else:
            return "F"
        
s1 = Student("Ali", 85)
print(s1.grade) # Output: B
```

<h3> Aggregation and Composition: <h3>

<p style="font-size:15px;"> Aggregation is a weak relationship between two classes where one class contains a reference to another class, but the contained class
can exist independently of the containing class. 
"I use you" <br>
Composition is a stronger form of aggregation where the contained class cannot exist independently of the containing class. 
In composition, the lifetime of the contained class is controlled by the containing class. If the containing class is destroyed, the contained class will also be destroyed.
"I own you" </p>

```python
# Aggregation:
class Library: # --> container or housing class for independant book class
    def __init__(self, name):
        self.name = name
        self.books = [] # --> You add books to a library outside of the class, and the books are not automatically created once library is constructed

    def add_book(self, book):
        self.books.append(book)

    def display_books(self):
        for i in self.books:
            print(i)

class Book:
    def __init__(self, title, author):
        self.title=title
        self.author=author

    def __str__(self):
        return f"{self.title} by {self.author}"

library = Library("The British Council Public Library")
b1 = Book("Harry Potter and the Philosophers Stone", "J.K. Rowling")
b2 = Book("Harry Potter and the Chamber of Secrets", "J.K. Rowling")
b3 = Book("Harry Potter and the Prisoner of Azkaban", "J.K. Rowling")

library.add_book(b1) # --> books created first then given to class
library.add_book(b2)
library.add_book(b3)
print(library.name)
library.display_books()

# Note: pass statement is a placeholder used when python expects code, but u havent written it yet, such as when u leave a class empty for future use. 
```

```python
# Composition:
class Room: # --> A room cant exist without a house, and once the house is destroyed the room will also no longer exist.
    def __init__(self, no_of_rooms):
        self.rooms = no_of_rooms

    def __str__(self):
        return f"{self.rooms}"

class House: # --> the house class owns the room class and room is made inside the house class.
    def __init__(self, location, size, no_of_rooms):
        self.location = location
        self.size = size
        self.rooms = Room(no_of_rooms) # --> when a house is created, python automatically creates room obj inside the class as house owns it.

    def __str__(self):
        return f"A house of {self.size} with {self.rooms} rooms at {self.location}"

h1 = House("Bahria Town", "650 sqft", "5")
print(h1)
```
